```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_fr = pd.read_csv("../data/FrenchOpen-women-2013.csv")
```

```python
df_us = pd.read_csv("../data/USOpen-women-2013.csv")
```

```python
df_us.rename(columns={"Player 1": "Player1", "Player 2": "Player2"}, inplace=True)
```

```python
def keep_last_name(df):
    df["Player1"] = df["Player1"].str.split(" ").str[-1]
    df["Player2"] = df["Player2"].str.split(" ").str[-1]
```

```python
keep_last_name(df_fr)
keep_last_name(df_us)
```

```python
df_merged1 = pd.merge(df_fr, df_us, on=["Player1", "Player2"], how="inner", suffixes=("_fr", "_us"))
```

```python
df_merged1
```

```python
df_merged2 = pd.merge(df_fr, df_us, left_on=["Player1", "Player2"], right_on=["Player2", "Player1"], how="inner", suffixes=("_fr", "_us"))
```

```python
df_merged2
```

```python
df_merged = pd.concat([df_merged1, df_merged2], ignore_index=True)
```
